```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Ryan Wong\Desktop\Beach_Water_Quality_-_Automated_Sensors.csv") #change this to the location of the data file
df
```

```python
#OBJECTIVES:
#check for inconsistancies with column names. make consistant if columns have same meaning. ex: 'ohio street beach' vs 'old - ohio street beach'
#split necessary columns into two seperate columns - timestamp column into DATE and TIME
#convert values of -99999 and -100000 into NaN
#delete irrelevant columns (Measurement Timestamp Label and Measurement ID - these two are repeats)


df['Beach Name'].value_counts() #view all unique column names
```

```python
df['Beach Name'].replace('Old - Ohio Street Beach', 'Ohio Street Beach', inplace = True) #replace 'old - ohio street beach' to 'ohio street beach'
df['Beach Name'].value_counts() #view unique column names to verify change - 'old - ohio street beach' should be deleted
```

```python
df[['Date', 'Time']] = df['Measurement Timestamp'].str.split(' ', 1, expand = True) #split 'measurement timestamp' column into two seperate columns - DATE and TIME
df['Date'] = pd.to_datetime(df['Date']) #convert date column to datetime format for code below to work
df['Month'] = df['Date'].dt.month #extracts month from date column
df['Year'] = df['Date'].dt.year #extracts year from date column
df
```

```python
df.drop(columns = ['Measurement Timestamp Label', 'Measurement ID'], inplace = True) #drop these two columns - repeated info
```

```python
#abnormal values discovered from above. convert these values to NaN. Need to use numpy library
#water temperature (-100000.0)
#turbidity (-100000.0)
#wave height (-99999.992)
#wave period (-100000.0) 

df['Water Temperature'].replace(-100000.0, np.nan, inplace = True)
df['Water Temperature'].value_counts() #verify abnormal values have been deleted
```

```python
df['Turbidity'].replace(-100000.0, np.nan, inplace = True)
df['Turbidity'].value_counts() #verify abnormal values have been deleted
```

```python
df['Wave Height'].replace(-99999.992, np.nan, inplace = True)
df['Wave Height'].value_counts() #verify abnormal values have been deleted
```

```python
df['Wave Period'].replace(-100000.00, np.nan, inplace = True)
df['Wave Period'].value_counts() #verify abnormal values have been deleted
#df[df['Wave Period'] < 0]['Wave Period'].value_counts()
```
