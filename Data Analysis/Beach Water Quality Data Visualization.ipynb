```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
df = pd.read_excel(r'C:\Users\Ryan Wong\Desktop\Beach Water Quality Cleaned.xlsx')
df
```

```python
#Do certain beaches have higher averages of measurements than others?
#How do the different months/seasons vary in terms of measurements than others? What about time of day? And years?
#What is the correlation to battery life to time of day/month or beach location?
#What is the correlation to water temperature and turbidity to wave height and wave period?

#Are there any extreme outliers in distribution of this dataset?
#How frequently is each beach measured?
#--------------------------------------

#Do certain beaches have higher averages of measurements than others?
#Measurements: Water Temperature, Turbidity, Transducer Depth
columnsToAnalyze = ['Water Temperature', 'Turbidity', 'Transducer Depth', 'Wave Height', 'Wave Period', 'Battery Life']

for column in columnsToAnalyze:
    current= df[['Beach Name', column]].groupby('Beach Name').median().sort_values(column, ascending=False)
    current.plot(kind='bar')
    plt.title(f'Median {column} to Beach Location')
    plt.xlabel('Beach Locations')
    plt.ylabel(f'Median {column}')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.75)
```

```python
import matplotlib.pyplot as plt

all = df.groupby('Beach Name')[['Water Temperature', 'Turbidity', 'Transducer Depth', 'Wave Height', 'Wave Period', 'Battery Life']].median().sort_values(by='Water Temperature', ascending=False)
plt.figure(figsize=(20,15))
all.plot(kind='bar')
plt.title('Median Measurements to Beach Locations')
plt.xlabel('Beach Locations')
plt.ylabel('Median Measurements')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.75)
```
