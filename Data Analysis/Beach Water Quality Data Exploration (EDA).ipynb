```python
#Do certain beaches have higher averages of measurements than others?

df.groupby('Beach Name').mean().sort_values(by='Water Temperature', ascending=False)
```

```python
#How do the different months/seasons vary in terms of measurements than others? What about time of day? And years?

#month/season:
df.groupby('Month')[['Water Temperature', 'Turbidity', 'Transducer Depth', 'Wave Height', 'Wave Period', 'Battery Life']].mean()
```

```python
#time of day:
df.groupby('Time')[['Water Temperature', 'Turbidity', 'Transducer Depth', 'Wave Height', 'Wave Period', 'Battery Life']].mean()
```

```python
#years:
df.groupby('Year')[['Water Temperature', 'Turbidity', 'Transducer Depth', 'Wave Height', 'Wave Period', 'Battery Life']].mean()
```

```python
#What is the correlation to battery life to time of day/month?
#what about to beach location?

df.groupby('Time')[['Battery Life']].mean().sort_values(by='Battery Life')
```

```python
#df.groupby('Month')[['Battery Life']].mean().sort_values(by='Battery Life')
df.groupby('Month').mean().sort_values(by='Battery Life')
```

```python
df.groupby('Beach Name')[['Battery Life']].mean().sort_values(by='Battery Life')
```

```python
#What is the correlation to water temperature and turbidity to wave height and wave period?

#Wave Period to Water Temperature
with pd.option_context('display.max_rows', None): #<-- display all rows instead of first and last 5
    print(df.groupby('Wave Period')[['Water Temperature']].mean().sort_values(by='Water Temperature'))
```

```python
#Wave Period to Turbidity
with pd.option_context('display.max_rows', None):
    print(df.groupby('Wave Period')[['Turbidity']].mean().sort_values(by='Turbidity'))
```

```python
#Wave Height to Water Temperature
with pd.option_context('display.max_rows', None):
    print(df.groupby('Wave Height')[['Water Temperature']].mean().sort_values(by='Water Temperature'))
```

```python
#Wave Height to Turbidity
with pd.option_context('display.max_rows', None):
    print(df.groupby('Wave Height')[['Turbidity']].mean().sort_values(by='Turbidity'))
```
